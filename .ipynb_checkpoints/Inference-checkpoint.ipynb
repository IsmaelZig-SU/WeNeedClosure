```python
import seaborn as sns
from tqdm import tqdm
import torch
import numpy as np
import matplotlib.pyplot as plt
from src.model import Diffusion_Transformer 
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
checkpoint = torch.load("Diffusion_exp_diffsteps100_emb_dim128_nblocks4_.pt", map_location=device)
```

```python
model = Diffusion_Transformer(**checkpoint['model_config'])
model.load_state_dict(checkpoint['model_state_dict'])
model.to(device)
model.eval()

schedule = checkpoint['scheduler_config']
alphas_cumprod = schedule['alphas_cumprod'].to(device)
betas = schedule['betas'].to(device)
alphas = schedule['alphas'].to(device)
T = schedule['T']

test_data = checkpoint['test_data']
test_rom = test_data['rom']
test_fom = test_data['fom']

trajs, D_rom, samples = test_rom.shape
_, D_fom, _ = test_fom.shape

test_rom_flat = test_rom.transpose(0, 2, 1).reshape(-1, D_rom)
test_fom_flat = test_fom.transpose(0, 2, 1).reshape(-1, D_fom)

print(f"ROM shape: {test_rom.shape}")
print(f"FOM shape: {test_fom.shape}")

print(f"Model loaded with {T} denoising steps.")
```

```python
rom_mean = np.var(test_rom_flat, axis = 0)
fom_mean = np.var(test_fom_flat, axis = 0)
plt.bar(np.arange(D_rom), rom_mean)
plt.bar(np.arange(D_fom), fom_mean)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(test_rom.ravel(), stat="density", kde=True, bins=100, ax=axes[0], color='skyblue')
axes[0].set_title("ROM Distribution")
axes[0].set_xlim(-0.15, 0.15)

sns.histplot(test_fom.ravel(), stat="density", kde=True, bins=100, ax=axes[1], color='salmon')
axes[1].set_title("FOM Distribution")
axes[1].set_xlim(-0.15, 0.15)

plt.tight_layout()
plt.show()
```

```python
plt.title('Noise/Denoise schedule')
plt.plot(betas.detach().cpu().numpy(), label = 'betas') 
plt.plot(alphas.detach().cpu().numpy(), label = 'alphas') 
plt.plot(alphas_cumprod.detach().cpu().numpy(), label = 'alphas cumprod') 
plt.legend()
plt.show()
```

```python
@torch.no_grad()
def sample(model, prior, shape, steps):
    """
    DDPM sampling for models predicting epsilon (noise).
    prior: The ROM conditioning signal [B, rom_dim]
    shape: shape of the sample [B, fom_dim]
    steps: number of diffusion steps
    """
    model.eval()
    device = prior.device

    x_t = torch.randn(shape, device=device)  # start from pure noise

    for t in tqdm(reversed(range(steps)), desc="Denoising"):
        beta_t = betas[t].to(device)
        alpha_t = alphas[t].to(device)
        alpha_bar_t = alphas_cumprod[t].to(device)

        # reshape for broadcasting
        beta_t = beta_t.view(1, 1)
        alpha_t = alpha_t.view(1, 1)
        alpha_bar_t = alpha_bar_t.view(1, 1)

        # predict epsilon
        eps_pred = model(x_t, prior)

        # posterior mean for x_{t-1}
        mean = (1 / torch.sqrt(alpha_t)) * (
            x_t - (1 - alpha_t) / torch.sqrt(1 - alpha_bar_t) * eps_pred
        )

        if t > 0:
            alpha_bar_prev = alphas_cumprod[t - 1].to(device).view(1, 1)
            var = beta_t * (1 - alpha_bar_prev) / (1 - alpha_bar_t)
            noise = torch.randn_like(x_t)
            x_t = mean + torch.sqrt(var) * noise
        else:
            x_t = mean

    return x_t
```

```python
traj = 0

low_res_batch = torch.tensor(
    test_rom[traj, :, :].transpose(1,0),#shape [snapshots, D_rom]
    dtype=torch.float32,
    device=device
) 

true_batch = torch.tensor(
    test_fom[traj, :, :].transpose(1,0),#shape [snapshots, D_fom]
    dtype=torch.float32,
    device=device
)

print(low_res_batch.shape)
sample_pred = sample(model, low_res_batch, true_batch.shape, T)
print(sample_pred.shape)
```

```python
x = low_res_batch.detach().cpu().numpy()
y = sample_pred.detach().cpu().numpy()
y_true = true_batch.detach().cpu().numpy()
snapshots = np.arange(samples)
print(x.shape, y.shape, y_true.shape)

plt.figure(figsize=(8,5))
plt.title(f'Modal distribution at t = {snapshots[-1]} snapshots')
plt.plot(x[-1, :], color = 'cornflowerblue', linewidth = 2, label = 'ROM modes [375]', alpha = 0.5) 
plt.plot(y_true[-1, :], color = 'coral', linewidth = 2, label = 'FOM modes [900]') 
plt.plot(y[-1, :], color = 'black', linewidth = 2, alpha = 0.8, linestyle = '--', label = 'Diffusion prediction [900]')
plt.legend()
plt.show()
```

```python
ens = 10
sample_ens = []

print(low_res_batch.shape, true_batch.shape)

for i in range(ens) : 
    sample_pred = sample(model, low_res_batch, true_batch.shape, steps)
    sample_ens.append(sample_pred.unsqueeze(0).detach().cpu().numpy())

sample_ens = np.concatenate(sample_ens, axis = 0)
y_mean = np.mean(sample_ens, axis = 0) 
y_std = np.std(sample_ens, axis = 0)


mode = 200
t = np.arange(y_mean.shape[0])

plt.figure(figsize=(8,5))
plt.fill_between(
    t,
    y_mean[:, mode] - 3*y_std[:, mode],
    y_mean[:, mode] + 3*y_std[:, mode],
    color="blue",
    alpha=0.15,
    label="3 std"
)

plt.fill_between(
    t,
    y_mean[:, mode] - y_std[:, mode],
    y_mean[:, mode] + y_std[:, mode],
    color="blue",
    alpha=0.35,
    label="1 std"
)

plt.plot(
    y_mean[:, mode],
    color="red",
    linewidth=2,
    label=f"Mean prediction (mode {mode})"
)

plt.plot(
    y_true[:, mode],
    color="black",
    linewidth=2,
    linestyle=":",
    label=f"Truth (mode {mode})"
)

plt.legend()
plt.xlabel("Time step")
plt.ylabel("Value")
plt.title(f"Prediction with uncertainty (mode {mode})")
plt.show()
```

```python
from scipy.stats import wasserstein_distance

wd_x = np.mean([wasserstein_distance(x[i], y_true[i]) for i in range(y_true.shape[0])])
wd_y = np.mean([wasserstein_distance(y[i], y_true[i]) for i in range(y_true.shape[0])])

plt.bar(['ROM vs FOM', 'Diffusion prediction vs FOM'], [wd_x, wd_y], color=['coral', 'cornflowerblue'])
plt.ylabel('Wasserstein Distance')
plt.title('Comparison of W-1 distance before and after diffusion')
plt.show()
```

```python
wd_x = np.array([wasserstein_distance(x[i], y_true[i]) for i in range(y_true.shape[0])])
wd_y = np.array([wasserstein_distance(y[i], y_true[i]) for i in range(y_true.shape[0])])

plt.figure(figsize=(10,5))
plt.plot(wd_x, label='ROM vs FOM', color='blue', marker='o', markersize=4)
plt.plot(wd_y, label='FOM (pred) vs FOM', color='orange', marker='s', markersize=4)
plt.xlabel('Timestep')
plt.ylabel('Wasserstein Distance')
plt.title('Wasserstein Distance over Timesteps')
plt.legend()
plt.grid(True)
plt.show()
```

```python
mse = np.sum((y - y_true)**2) / np.sum(y_true**2) * 100
print("Relative MSE (%):", mse)

rrmse_rows = np.sum((y - y_true)**2, axis=1) / np.sum(y_true**2, axis=1) * 100
plt.figure(figsize=(10,5))
plt.plot(rrmse_rows, marker='o', linestyle='-', color='orange')
plt.xlabel('Sample index')
plt.ylabel('Relative RMSE (%)')
plt.title('Relative RMSE per Sample')
plt.grid(True)
plt.show()
```
